```python
import pandas as pd
import numpy as np
```

```python
credits_df = pd.read_csv('tmdb_5000_credits.csv')
movies_df = pd.read_csv('tmdb_5000_movies.csv')
movies_df.head()
```

```python
credits_df['title'].head()
```

```python
credits_df.columns = ['id','title','cast','crew']
movies_df = movies_df.merge(credits_df,on="id")
movies_df
```

```python
from ast import literal_eval

features = ["cast","crew","keywords","genres"]

for feature in features:
    movies_df[feature] = movies_df[feature].apply(lambda x: literal_eval(str(x)))

movies_df[features].head()


    
```

```python
#To acquire director names from the list of movies
def get_director(x):
    for i in x:
        if i["job"]=='Director':
            return i["name"]
    return np.nan
```

```python
#To return a list
def get_list(x):
    if isinstance(x,list):
        names = [i["name"] for i in x]
        
        if len(names)>3:
            names = names[:3]
        
        return names
    
    return []
    
```

```python
movies_df["director"] = movies_df["crew"].apply(get_director)

features = ["cast","keywords","genres"]

for feature in features:
    movies_df[feature] = movies_df[feature].apply(get_list)

movies_df["director"].head()
```

```python
movies_df["cast"].head()
```

```python
movies_df.rename(columns = {'original_title':'title'}, inplace = True)

movies_df[['title','cast','director','keywords','genres']].head()
```

```python
movies_df[features]
```

```python
#Cleaning data (Remove spaces, lowercase etc)
def data_cleaner(row):
    if isinstance(row,list):
        return [str.lower(i.replace(" ",""))for i in row]
    else:
        if isinstance(row,str):
            return str.lower(row.replace(" ",""))
        else:
            return ""

        
features = ['cast','keywords','director','genres']
for feature in features:
    movies_df[feature] = movies_df[feature].apply(data_cleaner)
```

```python
movies_df[features]
```

```python
def vectorize(features):
    return ' '.join(features['keywords']) + ' ' + ' '.join(features['cast']) + ' ' + features['director'] + ' ' + ' '.join(features['genres'])
```

```python
movies_df["soup"] = movies_df.apply(vectorize,axis=1)
print(movies_df["soup"].head())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

count_vectorizer = CountVectorizer(stop_words='english')
count_matrix = count_vectorizer.fit_transform(movies_df["soup"])

print(count_matrix.shape)

cosine_simi = cosine_similarity(count_matrix,count_matrix)
print(cosine_simi.shape)

#Resetting index
movies_df = movies_df.reset_index()
indices = pd.Series(movies_df.index,index=movies_df['title'])
indices
```

```python
indices = pd.Series(movies_df.index,index=movies_df['title']).drop_duplicates() #Removing duplicates if any
print(indices)
```

```python
def get_recommendations(title,cosine_similarity):
    index = indices[title]
    similarityscore = list(enumerate(cosine_similarity[index]))
    similarityscore = sorted(similarityscore,key=lambda x: x[1],reverse=True)
    similarityscore = similarityscore[1:11] #Gives the first 10 movies with closest similarity scores apart from the movie itself 
    
    movies_indices = [ind[0] for ind in similarityscore]
    movies = movies_df['title'].iloc[movies_indices]
    return movies
```

```python
print(get_recommendations("Avatar",cosine_simi))
```

```python
print(get_recommendations("The Avengers",cosine_simi))
```
